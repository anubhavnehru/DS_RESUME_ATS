```python
import requests
import json
import csv
import pandas as pd
```

```python
# Read CSV data
csv_data = []
with open("cleaned_resume.csv", newline="") as csvfile:  # Replace with your CSV file
    reader = csv.reader(csvfile)
    next(reader) # Skip the header
    for row in reader:
        csv_data.append(row)

```

```python
# Prepare input for one-shot prediction
input_data = []
unique_job_titles = set()  # To keep track of unique job titles
for row in csv_data:
    job_title, resume = row[0], row[1]
    # If the job title is not already encountered, add it to the input data
    if job_title not in unique_job_titles:
        unique_job_titles.add(job_title)
        input_data.append({"job_title": job_title, "resume": resume})

```

```python
# Format input data as string
content = ""
for idx, entry in enumerate(input_data):
    content += f"\nExample {idx + 1}: Job Title: {entry['job_title']}\nResume: {entry['resume']}\n"
    
content
```

```python
import openai
import pandas as pd

# Set your Together AI API key
together_api_key = together_api_key

# Provided context data
context_data = content
# Initialize Together AI client
client = openai.OpenAI(api_key=together_api_key, base_url="https://api.together.xyz/v1")

# Function to predict job titles from resumes
def predict_job_titles(resumes, job_titles):
    predicted_titles = []
    for idx, resume in enumerate(resumes):
        print(f"Processing resume {idx + 1}")
        # Prompt for the model with job titles
        prompt = f"Please predict the job title based on the following resume:{resume} \
            Strictly select the job titles from the list: {', '.join(job_titles)}"

        # Send the request to the model
        response = client.chat.completions.create(
            messages=[{"role": "system", "content": context_data}, {"role": "user", "content": prompt}],
            model="mistralai/Mixtral-8x7B-Instruct-v0.1",
            max_tokens=50,
            temperature=0.1
        )

         # Extracting output
        output = response.choices[0].message.content.strip()
        # Extracting job title from the output
        output_job_title = output.split(":")[-1].strip()
        # Filtering the output based on provided job titles
        classified_job_titles = [title for title in job_titles if title.lower() 
                                 in output_job_title.lower()] or [output_job_title]
        print("Predicted Job Title:", classified_job_titles[0])
        predicted_titles.append(classified_job_titles[0])
    
    return predicted_titles
```

```python
# Example DataFrame with resumes and job titles

final_table = pd.read_csv('cleaned_resume.csv')

final_table1 = final_table.copy()
# Get unique job titles
job_titles = final_table['Category'].unique()

# Predict job titles for resumes
predicted_job_titles = predict_job_titles(final_table1['Cleaned_Resume'], job_titles)
```

```python
import re

# Function to extract job titles within double quotes
def extract_job_titles(text):
    # Use regular expression to find job titles within double quotes
    matches = re.findall(r'"(.*?)"', text)
    if matches :
        return matches[0]
    else : 
        return None 

predicted_category = mixtral_data['Predicted_Category']
```

```python
def check_same_category(category1, category2):
    # Convert both categories to lowercase for case-insensitive comparison
    og_category1 = category1
    og_category2 = category2
    
        
    category1 = category1.lower().replace(" " , "")
    category2 = category2.lower().replace(" ", "")
    
    # Check if one category is a substring of the other
    if category1 in category2 or category2 in category1:
        return og_category1
    else:
        return og_category2
```
